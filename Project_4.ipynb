```python
# College_statistics csv
college_statistics_path = Path(r"C:\Users\hsyuk\OneDrive\Desktop\GW Bootcamp\Section 5 - Advanced Topics (18-24)\Project_4\Resources\college_statistics.csv")
college_statistics_df = pd.read_csv(college_statistics_path)
college_statistics_df.head()
```

```python
# Drop alt_player_id, school_primary_color, school_alt_color columns
columns_to_drop = ['alt_player_id', 'school_primary_color', 'school_alt_color']
college_statistics_df = college_statistics_df.drop(columns=columns_to_drop, axis=1)
college_statistics_df.head()
```

```python
college_statistics_df.dropna()
```

```python
# ids csv
ids_path = Path(r"C:\Users\hsyuk\OneDrive\Desktop\GW Bootcamp\Section 5 - Advanced Topics (18-24)\Project_4\Resources\ids.csv")
ids_df = pd.read_csv(ids_path)
ids_df
```

```python
# Drop Espn_id
ids_df = ids_df.drop(columns = 'espn_id', axis = 1)
ids_df.head()
```

```python
ids_df.dropna()
```

```python
# nfl_draft_profiles csv
nfl_profiles_path = Path(r"C:\Users\hsyuk\OneDrive\Desktop\GW Bootcamp\Section 5 - Advanced Topics (18-24)\Project_4\Resources\nfl_draft_profiles.csv")
nfl_profiles_df = pd.read_csv(nfl_profiles_path)
nfl_profiles_df
```

```python
# drop columns
drop_columns = ['alt_player_id','link','school_logo','text1','text2','text3','text4']
nfl_profiles_df=nfl_profiles_df.drop(columns=drop_columns, axis=1)
nfl_profiles_df.head()
```

```python
nfl_profiles_df.dropna()
```

```python
# nfl_draft_prospects csv
nfl_prospects_path = Path(r"C:\Users\hsyuk\OneDrive\Desktop\GW Bootcamp\Section 5 - Advanced Topics (18-24)\Project_4\Resources\nfl_draft_prospects.csv")
nfl_prospects_df = pd.read_csv(nfl_prospects_path)
nfl_prospects_df
```

```python
# Drop columns player_image, school_abbr, link, team_logo_espn, team_abbr
columns_drop = ['player_image', 'school_abbr', 'link', 'team_logo_espn', 'team_abbr', 'grade', 'pos_rk', 'ovr_rk']
nfl_prospects_df= nfl_prospects_df.drop(columns = columns_drop, axis=1)
nfl_prospects_df
```

```python
# Merge 
nfl_merge_df = pd.merge(college_qbr_df, college_statistics_df, on=['player_name'])
# Join ids_df
nfl_merge_df = pd.merge(nfl_merge_df, ids_df, on=['player_name'])
#Join nfl_profiles
nfl_merge_df = pd.merge(nfl_merge_df, nfl_profiles_df, on=['player_name','guid'])
# Join nfl_prospects_df
nfl_merge_df = pd.merge(nfl_merge_df, nfl_prospects_df, on=['player_name','guid'])
nfl_merge_df
```
