```python
# nfl_draft_prospects csv
nfl_prospects_path = Path(r"C:\Users\hsyuk\OneDrive\Desktop\GW Bootcamp\Section 5 - Advanced Topics (18-24)\Project_4\Resources\nfl_draft_prospects.csv")
nfl_prospects_df = pd.read_csv(nfl_prospects_path)
nfl_prospects_df
```

```python
# Drop columns player_image, school_abbr, link, team_logo_espn, team_abbr
columns_drop = ['player_id','player_image', 'position','pos_abbr', 'school','school_name','team','trade_note','traded','school_abbr', 'link', 'team_logo_espn', 'team_abbr', 'grade', 'pos_rk', 'ovr_rk']
nfl_prospects_df= nfl_prospects_df.drop(columns = columns_drop, axis=1)
nfl_prospects_df
```

```python
nfl_prospects_df.dropna()
#nfl_prospects_df.drop_duplicates(subset=['player_name'], inplace=True)
```

```python
# Merge 
nfl_merge_df = pd.merge(college_qbr_df, college_statistics_df, on=['player_name'])
# Join ids_df
nfl_merge_df = pd.merge(nfl_merge_df, ids_df, on=['player_name'])
#Join nfl_profiles
nfl_merge_df = pd.merge(nfl_merge_df, nfl_profiles_df, on=['player_name','guid'])
# Join nfl_prospects_df
nfl_merge_df = pd.merge(nfl_merge_df, nfl_prospects_df, on=['player_name','guid'])
nfl_merge_df
```

```python
nfl_merge_df.dropna(inplace=True)
```

```python
nfl_merge_df=nfl_merge_df.drop(columns=['player_name','guid','player_id','statistic','nflfastR_id'])
```

```python
#nfl_merge_df['statistic']=nfl_merge_df['statistic'].astype(str)
```

```python
nfl_merge_df.info()
```

```python
# Separate the data into labels and features

# Seprate the y variable, the labels
y = nfl_merge_df['overall']
# Separate the X variable, the features
X= nfl_merge_df.drop(columns = ['overall'])
```

```python
y.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
```

```python
logistic_model = LogisticRegression(random_state=1)
logistic_model.fit(X_train,y_train)
```

```python
logistic_model_predict = logistic_model.predict(X_test)
```

```python
balanced_accuracy_score(y_test,logistic_model_predict)
```

```python
print(classification_report(y_test,logistic_model_predict))
```

```python
ros_model = RandomOverSampler(random_state = 1)

X_oversampled, y_oversampled = ros_model.fit_resample(X_train,y_train)
```

```python
y_oversampled.value_counts()
```

```python
logistic_ros = LogisticRegression(random_state = 1)

logistic_ros.fit(X_oversampled,y_oversampled)

logistic_ros_predict = logistic_ros.predict(X_test)
```

```python
balanced_accuracy_score(y_test,logistic_ros_predict)
```

```python
print(classification_report(y_test, logistic_ros_predict))
```
